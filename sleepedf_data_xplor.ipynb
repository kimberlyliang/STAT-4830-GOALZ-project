```python
import mne
import glob
import os
```

```python
base_dir = '/Users/tereza/spring_2025/STAT_4830/STAT-4830-GOALZ-project/data/sleep-edf-database-expanded-1.0.0'
```

```python
subfolders = ['sleep-cassette', 'sleep-telemetry']

# Initialize list to hold full paths of PSG files
psg_files = []

for sub in subfolders:
    # Build the search path for PSG files within this subfolder
    search_path = os.path.join(base_dir, sub, '*-PSG.edf')
    # Use glob to find all matching files
    files = glob.glob(search_path)
    psg_files.extend(files)

# Print out the found PSG files to verify the correct files are being located
print("Found PSG files:")
for f in psg_files:
    print(f)
```

```python
# Configuration: choose channels to use
use_multiple_channels = False  # Set True for EEG+EOG+EMG, False for single EEG
if use_multiple_channels:
    channels_to_load = ["EEG Fpz-Cz", "EOG horizontal", "EMG submental"]  # multi-channel
else:
    channels_to_load = ["EEG Fpz-Cz"]  # single-channel EEG (Fpz-Cz)

# Example file paths (to be adjusted to your dataset location)
psg_file = "SC4001E0-PSG.edf"         # example PSG file
hyp_file = "SC4001E0-Hypnogram.edf"   # corresponding hypnogram file

# Load only the selected channels from the PSG EDF
raw = mne.io.read_raw_edf(psg_file, include=channels_to_load, preload=True, verbose=False)
print("Loaded channels:", raw.ch_names)
```
